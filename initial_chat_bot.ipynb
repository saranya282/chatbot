```
"""This is a  program for chat bot...It will be a simple chat bot using console...
The tasks are as follows..
1. Greeting the user and asking for user's name
2.Wishing user properly based on time
3.Asking user how can bot help them and providing the choices for their help
4. Responding to user properly..."""

from datetime import datetime # importing datetime for date and time
import random # for making randommchoices
import pytz # for timezones import pytz library

def greet():
    greetings=["Hi,I'm alex. How can I help you and May I know your name please?","Wonderful,you are here, I am there to help you,May i Know your name?"]
    print(random.choice(greetings))
    # greeting the user from the greetings list


def time_Greeting():
    current_time = datetime.now(pytz.timezone('Asia/Kolkata'))
    # for local time 
    #greetings accoring to the time
    if current_time.hour >= 0 and current_time.hour < 5 :
        time_greeting="HI,I think you didn't have  a good sleep "
    elif current_time.hour >= 5 and current_time.hour < 12 :
        time_greeting="Good Morning"
    elif current_time.hour >= 12 and current_time.hour < 16 :
        time_greeting="Good Afternoon"
    elif current_time.hour >= 16 and current_time.hour <= 20 :
        time_greeting="Good evening"
    elif current_time.hour >= 20 and current_time.hour <= 24 :
        time_greeting="Hi,I think its time for sleep"
    print(time_greeting)




def welcome(name):
    welcome_wishes=["Glad to meet you","I hope we have great time together"]
    print(random.choice(welcome_wishes))

def show_choices():
    #giving choices to users from alex
    print("1.Calculate an expresssion: ")
    print("2.Date and time at greenwich meridian")
    print("3.Local Date and time ")
    print("4.A number or string  is palindrome or not")
    print("5.end the bot")
    print("------------------------")

    try:
        return int(input("Enter your choice[1-5]"))
    except:
        print("only enter choices from 1,2,3,4 and 5")
def evaluate_expression():
    expr=input("Enter an expression")
    try:
        print("Result = " , eval(expr))
        #for evaluating expressions
    except Exception as e:
        print(str(e))
def palindrome():
    #palindrome program
    n=input("enter something")
    if(n==n[::-1]):
        print("its a palindrome")
    else:
        print("not a palindrome")
    
def bot():
    
    greet()
    name=input("Your Name:")
    time_Greeting()


    welcome(name)
    choices = show_choices()
    while choices != 5 :
        if choices == 1 :
            evaluate_expression()
        elif choices == 2:
            print((datetime.now()))
        elif choices == 3 :
            print(datetime.now(pytz.timezone('Asia/Kolkata')))
        elif choices == 4 :
            palindrome()
        else:
            print("I dont understand that")# if number is other than 1,2,3,4
        choices = show_choices()
    
bot()


            





```
